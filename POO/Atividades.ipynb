### 1. Classe Bola: Crie uma classe que modele uma bola: 
### a. Atributos: Cor, circunferência, material
### b. Métodos: trocaCor e mostraCor

```python
class Bola():
  def __init__(self, cor, circ, material):
    self.cor = cor
    self.circ = circ
    self.material = material

  #metodo que troca a cor 
  def trocaCor(self, cor):
    self.cor = cor

  #metodo que imprime a cor 
  def mostraCor(self):
    return self.cor

#criando a bola1
bola1 = Bola("azul", "50", "plastico")

#imprimindo a cor da bola1
print(bola1.cor)

#trocando a cor da bola1
bola1.trocaCor("Verde")

#imprimindo a cor da bola1
print(bola1.mostraCor())
print(bola1.cor)
```

### 2. Classe Quadrado: Crie uma classe que modele um quadrado:
### a. Atributos: Tamanho do lado
### b. Métodos: Mudar valor do Lado, Retornar valor do Lado e calcular Área

```python
#versão 1.0

class Quadrado():
  def __init__(self, tam_lado):
    self.tam_lado = tam_lado
  
  def mudar_valor_do_lado(self, tam_lado):
    self.tam_lado = tam_lado

  def retornar_valor(self):
    return self.tam_lado

  def calcular_area(self):
    area = (int(self.retornar_valor()) ** 2)
    return area

quad1 = Quadrado("2")
quad1.mudar_valor_do_lado("4")
print(quad1.retornar_valor())
print(quad1.calcular_area())
print(quad1.tam_lado)
```

```python
#versão 2.0

class Quadrado():
  def __init__(self, tam_lado):
    #chama o metodo "mudar_valor_do_lado" para atribuir os valores
    self.mudar_valor_do_lado(tam_lado) 
  
  def mudar_valor_do_lado(self, tam_lado):
    self.tam_lado = tam_lado

  def retornar_valor(self):
    return self.tam_lado

  def calcular_area(self):
    return (self.retornar_valor()) ** 2

#criando um quadrado
quad1 = Quadrado(2)

#mudando o valor do lado
quad1.mudar_valor_do_lado(5)

#imprimindo o valor (tamanho do lado )
print(quad1.retornar_valor())

#imprimindo a area
print(quad1.calcular_area())

#imprimindo o tamanho do lado 
print(quad1.tam_lado)
```

### 3. Classe Pessoa: Crie uma classe que modele uma pessoa:
### a. Atributos: nome, idade, peso e altura
### b. Métodos: Envelhercer, engordar, emagrecer, crescer. 
### Obs: Por padrão, a cada ano que nossa pessoa envelhece, sendo a idade dela menor que 21 anos, ela deve crescer 0,5 cm.

```python
class Pessoa():
  #construtor
  def __init__(self, nome, idade, peso, altura):
    self.nome = nome
    self.idade = idade
    self.peso = peso
    self.altura = altura

  def envelhecer(self, idade):
    self.idade += idade
    if self.idade < 21:
      anos = idade - self.idade
      anos *= 0.5
      crescer(anos)
    
  def engordar(self, peso):
    self.peso += peso

  def emagrecer(self, peso):
    self.peso -= peso

  def crescer(self, altura):
    self.altura += altura

#criando pessoa1
pessoa1 = Pessoa("Maria", 30, 50.0, 1.65)

#pessoa1 engordou 5kg
pessoa1.engordar(5)

#imprimindo o pessoa da pessoa1
print(pessoa1.peso)

#pessoa1 envelheceu 5 anos
pessoa1.envelhecer(5)

#imprimindo a idade da pessoa1
print(pessoa1.idade)

#criando pessoa2
pessoa2 = Pessoa("José", 10 , 35, 1.40)

#pessoa 2 envelheceu 5 anos
pessoa2.envelhecer(5)

#imprimindo todos os atributos da pessoa2
print(vars(pessoa2))
```

### 4. Classe Conta Corrente: Crie uma classe para implementar uma conta corrente. A classe deve possuir os seguintes atributos: número da conta, nome do correntista e saldo. Os métodos são os seguintes: alterarNome, depósito e saque; No construtor, saldo é opcional, com valor default zero e os demais  atributos são obrigatórios.

```python
class Conta_corrente():
  #construtor
  def __init__(self, num_conta, nome_correntista, saldo):
    self.num_conta = num_conta
    self.nome_correntista = nome_correntista
    self.saldo = saldo

  def altera_nome(self, nome):
    self.nome_correntista = nome
    
  def deposito(self, deposito):
    self.saldo += deposito
    
  def saque(self, saque):
    self.saldo -= saque

#criando um conta
conta1 = Conta_corrente(1010, "Maria", 300.0)

#imprimindo todos os atributos da conta
print(vars(conta1))

#realizando o saque de R$100,00
conta1.saque(100)

#imprimindo todos os atributos da conta
print(vars(conta1))

#realizando o depósito de R$1000,00
conta1.deposito(1000)

#imprimindo todos os atributos da conta
print(vars(conta1))
```
