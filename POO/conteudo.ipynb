```python
class Veiculo():
  def __init__(self, tipo, chassi, marca, modelo, ano):
    self.tipo = tipo
    self.chassi = chassi
    self.marca = marca
    self.modelo = modelo
    self.ano = ano

class Motocicleta(Veiculo):
  def __init__(self, tipo, chassi, marca, modelo, ano, cilindrada):
    super().__init__(tipo, chassi, marca, modelo, ano)
    self.cilindrada = cilindrada

v = Veiculo('carro', '5454d34dfds', 'Ford', 'KA','2012')
print(vars(v))

m = Motocicleta('motocicleta', '5464vdvd', 'Honda', 'CG','2012', 150)
print(vars(m))
```

```python
class Funcionario:
  def __init__(self, nome, cpf, salario):
    self.nome = nome
    self.cpf = cpf
    self.salario = salario

  #polimorfismo
  def get_bonificacao(self):
    return self.salario * 0.15

class Gerente(Funcionario):
  def __init__(self, nome,cpf, salario, senha, qtd_funcionarios):
    super().__init__(nome,cpf, salario)
    self.senha = senha
    self.qtd_funcionarios = qtd_funcionarios

  def get_bonificacao(self):
    #está puxando o metodo da classe Funcionario 
    return super().get_bonificacao() * 2

joao = Funcionario("joão", "5464654546545", 1000)
heloise = Gerente("hELOISE0", "876884546", 1000, "ADMIN", 54)

print(vars(joao))
print(joao.get_bonificacao())
print(vars(heloise))
print(heloise.get_bonificacao())

```

### classe abstrata

classe abstrata não é para sofrer alteração (mas pode)

```python
from abc import ABC, abstractmethod

class letras():
  @abstractmethod #para mostrar que a classe é abstrata
  
  def mostrarTipo(self):
    print("Eu sou uma classe abstrata!")
  
class A(letras):
  
  def __init__(self, descricao):
    self.descricao = descricao

  def imprimir(self):
    print("Aqui é um método da classe A!")

exemplo_Letra = A("Letra A")
print(vars(exemplo_Letra))

exemplo_Letra.mostrarTipo()
exemplo_Letra.imprimir()
```

###Classe carro: Implemente uma classe chamada Carro com as seguintes propriedades:
###a. Um veículo tem um certo consumo de combustível (medidos em km / litro) e uma certa quantidade de combustível no tanque.
###b. O consumo é especificado no construtor e o nível de combustível inicial é 0.
###c. Forneça um método andar( ) que simule o ato de dirigir o veículo por uma certa distância, reduzindo o nível de combustível no tanque de gasolina.
###d. Forneça um método obterGasolina( ), que retorna o nível atual de combustível.
###e. Forneça um método adicionarGasolina( ), para abastecer o tanque. Exemplo de uso:

```python
class Carro():
  def __init__(self, consumo_combus):
    self.consumo_combus = consumo_combus#km/L
    self.qtd_combus = 0
  
  def andar(self, distancia):
    qtd_litros = distancia / self.consumo_combus
    self.qtd_combus -= qtd_litros
  
  #retorna o nível atual de combustível
  def obterGasolina(self):
    return self.qtd_combus
  
  #abastecer o tanque
  def adicionarGasolina(self, add):
    self.qtd_combus += add


carro1 = Carro(7)#km/l  | Litros
print(vars(carro1))
#carro1 = Carro(7,10)#km/l  | Litros
carro1.adicionarGasolina(10)
print(vars(carro1))
carro1.andar(7)
print(vars(carro1))
print(carro1.obterGasolina())
```
