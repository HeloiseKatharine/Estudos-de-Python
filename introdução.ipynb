#Operadores aritiméticos

```python
n1 = int(input('Digite um valor: '))
n2 = int(input('Digite um valor: '))

#soma
soma = n1 + n2

#subtração
sub = n1 - n2

#divisão
div = n1 / n2

#divisão inteira
div_inteira = n1 // n2

#resto da divisão 
resto_divisão = n1 % n2

#potência
potencia = n1**n2

print(f'\n{soma}\n{sub}\n{div}\n{div_inteira}\n{resto_divisão}\n{potencia}')
```

# Operadores de atribuição

```python
x = 10

x += 10 (x = x + 10)

x -= 10 (x = x - 10)

x *= 10 (x = x * 10)

x /= 10 (x = x / 10)

x %= 10 (x = x % 10)

x **= 10 (x = x ** 10)

x //= 10 (x = x // 10)
```
