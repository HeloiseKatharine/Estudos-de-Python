#Tuplas

* Tuplas são imutáveis (Não pode alterar os valores)
* Não pode adicionar um item em uma tupla, pois ela é imutável.
* Não pode excluir um item de uma tupla, pois ela é imutável.





### Declaração de tuplas

```python
tupla = 'a', 'b', 'c', 'd'

#mais utilizada
tupla1 =  ('a', 'b', 'c') 

#tem que ter a vírgula depois 
tupla2 = ('a',) 

#sem a vírgula é str e não tupla
tupla3 = ('a') 

#tupla vazia
tupla4 = () 

tupla5 = ('a', 'b', 'c', 1, 2, 3) #pode ter vários tipos de dados  

print(tupla)
print(type(tupla))
print(tupla1)
print(type(tupla2))
print(type(tupla3))
print(tupla4)
print(tupla5)

```

###Imprimindo valores da tupla

```python
tupla =  ('a', 'b', 'c') 

print(tupla[0])
print(tupla[1])
print(tupla[2])
```

```python
frutas = ('banana', 'abacate', 'jabuticaba')

for nome in frutas:
  print('O nome é:', nome)
```

###Tamanho da tupla

```python
frutas = ('banana', 'abacate', 'jabuticaba')
print(len(frutas))
```

###Slicing

```python
frutas = ('banana', 'abacate', 'jabuticaba')
print(frutas[1:])
print(frutas[0:2])
```

###Verificando o índice  dos elementos da tupla

```python
frutas = ('banana', 'abacate', 'jabuticaba')
print(frutas.index('jabuticaba'))
```

###Deletando a tupla (del)

```python
frutas = ('banana', 'abacate', 'jabuticaba')
del frutas
```

###Convertendo uma tupla para lista

```python
frutas = ('banana', 'abacate', 'jabuticaba')
#tupla para lista
lista_frutas = list(frutas)

lista_frutas.append('manga')

#lista para tupla
frutas = tuple(lista_frutas)

print(lista_frutas)
print(frutas)
```
