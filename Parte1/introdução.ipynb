#Operadores aritiméticos

```python
n1 = int(input('Digite um valor: '))
n2 = int(input('Digite um valor: '))

#soma
soma = n1 + n2

#subtração
sub = n1 - n2

#divisão
div = n1 / n2

#divisão inteira
div_inteira = n1 // n2

#resto da divisão 
resto_divisão = n1 % n2

#potência
potencia = n1**n2

print(f'\n{soma}\n{sub}\n{div}\n{div_inteira}\n{resto_divisão}\n{potencia}')
```

### Manipulando string

* String é imutável

```python
frase = "testando manipulação de string"

#retorna a letra do índice 9 
print(frase[9])

#do índice 9 até o 20
print(frase[9:21])

#do 9 até o 21 pulando de 2 em 2
print(frase[9:21:2])

#começa do início (zero) e vai até o 5 
print(frase[:5])

#começa no 15 e vai até o final
print(frase[15:])

#começa no 9 e vai até o final e pula de 3 em 3 
print(frase[9::3])

#trás para frente 
print(frase[::-1])

#retorna tudo menos a última letra
print(frase[:-1])

#retorna tudo 
print(frase[:])

#última letra
print(frase[-1])

#retorna os elementos pulando de um em um 
print(frase[::1])

#retorna os elementos pulando de dois em dois
print(frase[::2])
```
